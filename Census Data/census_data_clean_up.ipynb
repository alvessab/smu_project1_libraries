```python
import pandas as pd
```

```python
file_path = "census_data.csv"

# Read the census data
census_df = pd.read_csv(file_path)
census_df.head()
```

```python
census_df.info()
```

```python
census_df.describe()
```

```python
#get rid of populations with less 0 individuals (census data places -66666666.0 as a null value)
mask= census_df.total_population >=0
census_df2= census_df.loc[mask]
census_df2
```

```python
#number of rows dropped
264960-251091
```

```python
census_df2.describe()
```

```python
#get rid of "null" values from median_age
mask = census_df2.median_age >=0
census_df3= census_df2.loc[mask]
census_df3
```

```python
#number of rows dropped
251091-251088
```

```python
#check for any odd values in columns of interest
census_df3.describe()
```

```python
#Save as a csv
census_df3.to_csv("census_data_cleaned.csv", encoding="utf-8", index=False)
```
