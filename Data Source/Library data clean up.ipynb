```python
import pandas as pd
```

```python
file_path = "Libraries.csv"

# Read the mouse data and the study results
library_data = pd.read_csv(file_path)
library_data.head()
```

```python
library_data.columns
```

```python
library_df= library_data[["State", 'Library ID', 'Submission Year', 'Library Name',
       'Street Address', 'City', 'Zip Code', 'Longitude', 'Latitude',
       'State Code', 'County Code', 'County', 'County Population',"Print_Collection_Expenditures",
       'Digital_Collection_Expenditures', 'Total Collection Expenditures','Total Operating Expenditures', 'Print Collection', 'Digital Collection', 'Audio Collection','Library Visits',
        'Registered Users']]
library_df
```

```python
library_df.info(0)
```

```python
mask= library_df["Print_Collection_Expenditures"] !="H_14"
library_df= library_df.loc[mask]
```

```python
#convert columns to floats

columns= ["Print_Collection_Expenditures",'Digital_Collection_Expenditures',
       'Total Collection Expenditures','Total Operating Expenditures', 'Print Collection', 'Digital Collection', 'Audio Collection','Library Visits',
        'Registered Users']

for col in columns:
    library_df[col] = library_df[col].astype(float)
    print(col)
library_df.info()
```

```python
#Save as a csv
library_df.to_csv("library_df.csv", encoding="utf-8", index=False)
```
