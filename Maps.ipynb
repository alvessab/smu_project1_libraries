```python
# Configure gmaps with API key
gmaps.configure(api_key=gkey)
```

```python
#read in files
file_path = "Merge Data Folders/official.csv"

# Read the census data 
df = pd.read_csv(file_path)
df.head()
```

```python
df.info()
```

```python
#Calculate our Registered User to Total Population Percentage for each row
df["reg_users_perc"] = (df["Registered Users"] / df["total_population"])
df.head()
```

```python
mask = df["Registered Users"] >= 0
data = df.loc[mask]
data
```

```python
mask = data["Registered Users"] <= data["total_population"]
data1 = data.loc[mask]
data1
```

```python
print(data1["Registered Users"].min())
print(data1["Registered Users"].max())
```

```python
registered_users = data1["reg_users_perc"]
```

```python
locations = data1.loc[:, ["Latitude", "Longitude"]]

# Create a poverty Heatmap layer
fig = gmaps.figure()

heat_layer = gmaps.heatmap_layer(locations, weights=registered_users, 
                                 dissipating=False, max_intensity=100,
                                 point_radius = 1)

fig.add_layer(heat_layer)

fig
```
